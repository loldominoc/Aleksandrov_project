## Контрольні запитання

1. У чому відмінність інтерпретатора від компілятора? Python -- компілятор чи інтерпретатор?
    
    `Інтерпретатор` виконує програму построково, без попереднього перетворення всього коду в машинний код. Кожна команда виконується відразу після прочитання.
    
    `Компілятор` перетворює весь вихідний код програми в машинний код (бінарний файл), який потім виконується.

1. Що таке IDE PyCharm?

    `PyCharm` — це інтегроване середовище розробки (IDE) для Python. Воно надає інструменти для редагування, запуску та налагодження програм, включає в себе автозаповнення коду, підсвічування синтаксису, контроль версій, інтеграцію з системами керування базами даних та інші функції для підвищення ефективності розробки.

1. Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ SVN та GIT?

    `Система контролю версій (СКВ)` — це інструмент, що дозволяє зберігати і відстежувати зміни в коді, а також керувати різними версіями проєктів. Основна мета СКВ — дозволити командам працювати над проєктами спільно, надаючи можливість відстежувати зміни, повертатися до попередніх версій та уникати конфліктів при одночасній роботі декількох розробників.

    + `SVN` — централізована СКВ. Є один центральний сервер, на якому зберігається весь код, і всі користувачі працюють з ним безпосередньо.
    + `GIT` — децентралізована СКВ. Кожен користувач має свою локальну копію репозиторію, і всі зміни можуть бути синхронізовані через віддалений репозиторій (наприклад, на GitHub).

1. Що відбувається при виконанні команди `git add .`?

    Команда `git add .` додає всі файли в поточній директорії (включаючи піддиректорії) до індексу (staging area), тобто до "черги" для подальшого коміту. Це підготовка до збереження змін у репозиторії.

1. Що відбувається при виконанні команди `git commit`?

    Команда `git commit` зберігає всі зміни, додані до індексу, у вигляді нової версії (коміту) в локальному репозиторії. При цьому потрібно додати повідомлення, яке описує, що було зроблено в цьому коміті.

1. Якою Git-командою виконується розміщення локальної копії Git-репозиторію проєкта на віддаленому, що знаходиться на GitHub?

    Для розміщення локальної копії на віддаленому сервері GitHub використовується команда: `git push origin <назва_гілки>` та `git push origin main`
